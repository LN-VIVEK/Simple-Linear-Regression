```python
# STEP 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# STEP 2: Import Data File
dataset = pd.read_csv("RCC Cost check.csv")

# Optional: Clean column names (important)
dataset.columns = ["RCC_Area", "RCC_Cost"]

# Step 3: View Data
print(dataset.head())

# Step 4: Input / Output Split
X = dataset[["RCC_Area"]]   # Independent variable
Y = dataset[["RCC_Cost"]]   # Dependent variable

# Step 5: Data Visualization (optional)
plt.scatter(X, Y)
plt.xlabel("RCC Area (sq.m)")
plt.ylabel("RCC Cost (INR)")
plt.title("RCC Cost vs Area")
plt.show()

# Step 6: Split Train & Test
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=0
)

# Step 7: Model Creation
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, Y_train)

# Step 8: Weight & Bias
weight = regressor.coef_
bias = regressor.intercept_

print("\nWeight (Slope):", weight)
print("Bias:", bias)

# Step 9: Prediction
Y_pred = regressor.predict(X_test)

print("\nPredicted Values:")
print(Y_pred)

# Step 10: Evaluation (R² Score)
from sklearn.metrics import r2_score

r_score = r2_score(Y_test, Y_pred)
print("\nR² Score:", r_score)

# STEP 11: Save Model
import pickle

filename = "rcc_model.sav"
pickle.dump(regressor, open(filename, "wb"))

print("\n✅ Model Saved as rcc_model.sav")

# STEP 12: Load Model
loaded_model = pickle.load(open("rcc_model.sav", "rb"))

# STEP 13: Prediction (User Input)
area = float(input("\nEnter RCC Area (sq.m): "))

result = loaded_model.predict([[area]])

print("✅ Predicted RCC Cost =", result[0][0])
```
