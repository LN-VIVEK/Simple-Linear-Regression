```python
# STEP 1: Import Libraries
import pandas as pd 
import matplotlib.pyplot as plt

```

```python
#STEP 2: Import Dataset
dataset = pd.read_csv("Cricket.csv")
```

```python
# STEP 3: View Dataset
print(dataset.head())
```

```python
# STEP 4: Input / Output Split
X = dataset[["Over"]]                # Independent variable
Y = dataset[["Cumulative Run"]]      # Dependent variable
```

```python
# STEP 5: Visualize Data
plt.scatter(X, Y)
plt.xlabel("Over")
plt.ylabel("Cumulative Run")
plt.title("Cricket Data - Over vs Runs")
plt.show()
```

```python
# STEP 6: Split Train & Test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
# STEP 7: Model Creation
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
# STEP 8: Get Weight & Bias
weight = regressor.coef_
bias = regressor.intercept_
print("\nWeight (Slope) =", weight)
print("Bias =", bias)
```

```python
# STEP 9: Prediction (Test Data)
Y_pred = regressor.predict(X_test)
print("\nPredicted Values:")
print(Y_pred)
```

```python
# STEP 10: Evaluation
from sklearn.metrics import r2_score
r_score = r2_score(Y_test, Y_pred)
print("\nR² Score =", r_score)
```

```python
# STEP 11: Save Model
import pickle
filename = "cricket_model.sav"
pickle.dump(regressor, open(filename, "wb"))
print("\n✅ Model Saved!")
```

```python
# STEP 12: Load Model
loaded_model = pickle.load(open("cricket_model.sav", "rb"))
```

```python
# STEP 13: User Prediction
over = float(input("\nEnter Over number: "))
result = loaded_model.predict([[over]])
print("✅ Predicted Cumulative Runs =", result[0][0])
```
